```python
# Reading Text Files Example

f = open('Trump Speech.txt','r')
message = f.read()
print(message)
f.close()
```

```python
# FreqDist function from NLTK can be used to obtain word counts
# There are several other methods as well

import nltk
from nltk.probability import FreqDist

fdist1 = FreqDist(nltk.tokenize.word_tokenize(message))
```

```python
# 
fdist1.most_common(50)
```

```python
import csv
```

```python
import sys

import codecs

with open('Trump_speech.csv', 'w', encoding='utf8',newline='') as csvfile:
    wf = csv.writer(csvfile, delimiter=',',
                            quotechar='"')
    
    wf.writerow(["Words","Frequency"])
    for it in fdist1.keys():
        wf.writerow([it, fdist1[it]])
```
